```python
# Definindo funções e as derivadas numéricas.
derivada_num_avancada = lambda f, x, h: (f(x + h) - f(x))/h
derivada_num_centrada = lambda f, x, h: (f(x + h)-f(x - h))/(2 * h)
derivada_num_atrasada = lambda f, x ,h: (f(x) - f(x - h))/h
derivada_num_lagrange = lambda f, x, h, h2: ((-2 * h - h2)/(h**2 + h * h2)) * f(x) + ((h + h2)/(h * h2)) * f(x + h) + ((-h)/(h2**2 + h * h2)) * f(x + h + h2)

f = lambda x: np.sin(x)**2
g = lambda x: (x**5) - (x**4)

df = lambda x: (2 * np.sin(x)*np.cos(x))
dg = lambda x: (5 * (x**4) - 4 * (x**3))

def _derivadas_numericas_(f, x, h, h2, metodo):
    if(metodo == 'avancada'):
        return (f(x + h) - f(x))/h
    elif(metodo == 'centrada'):
        return (f(x + h) - f(x - h))/(2 * h)
    elif(metodo == 'atrasada'):
        return (f(x) - f(x - h))/h
    elif(metodo == 'lagrange'):
        return ((-2 * h - h2)/(h**2 + h * h2)) * f(x) + ((h + h2)/(h * h2)) * f(x + h) + ((-h)/(h2**2 + h * h2)) * f(x + h + h2)
    else:
        return "Este método não esta presente..."

def plot_atrasada(f, df, x, h, label):
    sol_exata = df(x)
    diferenca_approximada = _derivadas_numericas_(f, x, h, 0, "avancada")
    # Plot do resultado no gráfico
    plt.figure(dpi=100, figsize = (8, 6))
    plt.title(label=label+str(h), fontdict=None, loc="center", pad=None)
    plt.plot(x, sol_exata, '--' ,label="Solução exata")
    plt.plot(x, diferenca_approximada, '-' ,label="Solução aproximada")
    plt.legend()
    plt.show()
    # retorno do erro
    return max(abs(sol_exata-diferenca_approximada))

def plot_centrada(f, df, x, h, label):
    sol_exata = df(x)
    diferenca_approximada = _derivadas_numericas_(f, x, h, 0, "centrada")
    # Plot do resultado no gráfico
    plt.figure(dpi=100, figsize = (8, 6))
    plt.title(label=label+str(h), fontdict=None, loc="center", pad=None)
    plt.plot(x, sol_exata, '--' ,label="Solução exata")
    plt.plot(x, diferenca_approximada, '-' ,label="Solução aproximada")
    plt.legend()
    plt.show()
    # retorno do erro
    return max(abs(sol_exata-diferenca_approximada))

def plot_avancada(f, df, x, h, label):
    sol_exata = df(x)
    diferenca_approximada = _derivadas_numericas_(f, x, h, 0, "atrasada")
    # Plot do resultado no gráfico
    plt.figure(dpi=100, figsize = (8, 6))
    plt.title(label=label+str(h), fontdict=None, loc="center", pad=None)
    plt.plot(x, sol_exata, '--' ,label="Solução exata")
    plt.plot(x, diferenca_approximada, '-' ,label="Solução aproximada")
    plt.legend()
    plt.show()
    # retorno do erro
    return max(abs(sol_exata-diferenca_approximada))

def plot_lagrange(f, df, x, h1, h2, label):
    sol_exata = df(x)
    diferenca_approximada = _derivadas_numericas_(f, x, h1, h2, "lagrange")
    # Plot do resultado no gráfico
    plt.figure(dpi=100, figsize = (8, 6))
    plt.title(label=label+"h1="+str(h1)+" h2="+str(h2), fontdict=None, loc="center", pad=None)
    plt.plot(x, sol_exata, '--' ,label="Solução exata")
    plt.plot(x, diferenca_approximada, '-' ,label="Solução aproximada")
    plt.legend()
    plt.show()
    # retorno do erro
    return max(abs(sol_exata-diferenca_approximada))
```

```python
print("=====================================================================================================\n")
print("** Integração numérica **")
# Funções.

# Métodos vistos em aula.
ponto_medio = lambda f, a, b: f((a + b)/2) * (b - a)
trap = lambda f, a, b: (f(b) + f(a))*(b - a)/2
simpson = lambda f, a ,b, c: (f(a) + 4 * f(c) + f(b)) * ((b - a)/6)

# Métodos compostos
def composto_ponto_medio(f, a, b, h, n):
    h = float(b - a)/n
    integration = 0
    for i in range(n-1):
        integration = integration + f((a + h/2.0) + i * h)
    integration = integration * h
    return integration

def composto_trap(f, a, b, h, n):
    h = (b - a)/float(n)
    integration = f(a) + f(b)
    for i in range(1, n-1, 1):
        integration = integration + 2*f((a + i * h))
    integration = integration * h/2.0
    return integration

def composto_simpson(f, a, b, h, n):
    #if(n % 2 is not 0):
    #    return None
    h = (b - a)/float(n)
    integration = f(a) + f(b)
    for i in range(1, n-1, 1):
        k = a + (i * h/2)
        if(i % 2 == 0):
            integration += 2 * f(k)
        else:
            integration += 4 * f(k)
    integration = integration * h/3.0
    return integration

def _metodos_(f, fdx, a, b, tolerancia, n, metodo):
    vetor = []
    computado = 0
    diferenca = 0
    esperado = (fdx(b) - fdx(a))
    while(True):
        if(metodo == "ponto_medio"):
            computado = composto_ponto_medio(f, a, b, tolerancia, n)
        elif(metodo == "trapezio"):
            computado = composto_trap(f, a, b, tolerancia, n)
        elif(metodo == "simpson"):
            computado = composto_simpson(f, a, b, tolerancia, n)
        else:
            print("Método não definido...")
            return None
        diferenca = abs(esperado - computado)
        vetor.append(diferenca)
        if(diferenca < tolerancia):
            break
        n += 2
    print("Método " + metodo + "\n")
    print("Esperado: ", esperado)
    print("Computado:", computado)
    print("Diferença: ", diferenca)
    print("Sub divisões necessárias: ", n)
    print("\n")
    return vetor

# Métodos compostos

# Funções.
print("=====================================================================================================\n")
```
