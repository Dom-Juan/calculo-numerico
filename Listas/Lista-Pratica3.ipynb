```python
import numpy as np
import matplotlib.pyplot as plt

from math import pi as PI
from math import sin as SENO
from math import cos as COS
from tabulate import tabulate   # Import da biblioteca para facilitar o print de tabelas.
```

```python
# Algoritmos a serem usados
def metodo(function, a, b, t, iteracao_max):
    for i in range(iteracao_max):
        m = (a+b)/2
        if(abs(function(m)) < t):
            break
        if(function(a)*function(m) < 0):
            b = m
        else:
            a = m
    return m

def bisseccao_log(function, a, b, tolerancia, iteracao_max):
    i = 0
    x = (b+a)/2
    iteracao = list()
    afunction = list()
    bfunction = list()
    vetor_x = list()
    func = list()
    distancia = list()
    while(i < iteracao_max and function(x) != 0 and (b-a)/2 > tolerancia):
        if(np.sign(function(a)) == np.sign(function(b))): # function(a) * function(b) > 0
            a = x
        else:
            b = x
        iteracao.append(i+1)
        afunction.append([np.round(a,decimals=4),np.round(function(a),decimals=4)])
        bfunction.append([np.round(b,decimals=4),np.round(function(b),decimals=4)])
        vetor_x.append(np.round(x,decimals=4))
        func.append([np.round(x,decimals=4),np.round(function(x),decimals=4)])
        distancia.append((b-a)/2)
        x = (b+a)/2
        i = i + 1
    tabela = {
        'Método de bisseccao':["*" for i in range(len(iteracao))],
        'Iteração':[iteracao[i] for i in range(len(iteracao))],
        'a=a-f(a)':[afunction[i] for i in range(len(afunction))],
        'b=b-f(b)':[bfunction[i] for i in range(len(bfunction))],
        'x':[vetor_x[i] for i in range(len(vetor_x))],
        'f(x)':[func[i] for i in range(len(func))],
        'Distancia':[distancia[i] for i in range(len(distancia))]
        }
    print(tabulate(tabela, headers="keys", tablefmt='fancy_grid'))
    return x, tabela

def ponto_fixo(function, x, t, iteracao_max):
    for k in range(iteracao_max):
        x_ = function(x)
        if(abs(x - x_) < t):
            x = x_
            break
        x = x_
    print(x)
    return x

def newton_metodo(x, x_, function, dfunction, e_ant, n):
    iteracao = list()
    vetor_x = list()
    erro = list()
    erro_div = list()
    func = list()
    for i in range(n):
        x = x - function(x)/dfunction(x)
        e = np.abs(x_ - x)
        iteracao.append(i)
        vetor_x.append(x)
        erro.append(e)
        erro_div.append(e/e_ant**2)
        func.append(function(x))
        e_ant = e
    tabela = {
            'Método de Newton':["*" for i in range(len(iteracao))],
            'Iteração':[iteracao[i] for i in range(len(iteracao))],
            'x':[vetor_x[i] for i in range(len(vetor_x))],
            'f(x)':[func[i] for i in range(len(func))],
            'e=|x - x_ant|':[erro[i] for i in range(len(erro))],
            'Erro/Erro_anterior':[erro_div[i] for i in range(len(erro_div))],
        }
    print(tabulate(tabela, headers="keys", tablefmt='fancy_grid'))
    return tabela

def newton_metodo2(x, x_, function, dfunction, t, iteracao_max):
    iteracao = list()
    vetor_x = list()
    vetor_x_ant = list()
    erro_ = list()
    func = list()
    k = 0
    x_ant = 0
    #Método passado em aula
    while(k < iteracao_max and (np.abs(x_ant - x) > t or function(x) > t)):
        x_ant = x
        x = x_ant - function(x_ant)/dfunction(x_ant)
        iteracao.append(k)
        vetor_x.append(x)
        vetor_x_ant.append(x_ant)
        erro_.append(np.abs(x_ant - x))
        func.append(function(x))
        k = k + 1
    tabela = {
        'Método de Newton':["*" for i in range(len(iteracao))],
        'Iteração':[iteracao[i] for i in range(len(iteracao))],
        'x':[vetor_x[i] for i in range(len(vetor_x))],
        'x_anterior':[vetor_x_ant[i] for i in range(len(vetor_x_ant))],
        'f(x)':[func[i] for i in range(len(func))],
        'e=|x - x_ant|':[erro_[i] for i in range(len(erro_))],
        }
    print(tabulate(tabela, headers="keys", tablefmt='fancy_grid'))
    return tabela
```

```python
a = 2
b = PI/4
c = -2.5
d = PI/4

a_ = np.sqrt(4)
b_ = np.sqrt((1+SENO(PI/4)))
c_ = np.sqrt(np.abs(3*-2.5**2+6*-2.5))
d_ = np.sqrt((-0.2*COS(PI/4)+1))

# Funções escritas
fa = lambda x: x**3-2*x**2
dfa = lambda x: 3*x**2-4*x

fb = lambda x: x-COS(x)
dfb = lambda x: 1+SENO(x)

fc = lambda x: x**3+3*x**2-1
dfc = lambda x: 3*x**2+6*x

fd = lambda x: x-0.8-0.2*SENO(x)
dfd = lambda x: -0.2*COS(x)+1

tolerancia = 1E-3
iteracao_max = 120
print("Exemplos:\n")
print("a) x^3 - 2*x^2")
print("b) x - cos(x)")
print("c) x^3 + 3x^2 - 1")
print("d) x - 0.8 - 0.2*sen(x)\n")

print("\nExercício 1) Use o método de Newton para resolver os exemplos")
tabela_newton_a = newton_metodo2(a, a_, fa, dfa, tolerancia, iteracao_max)
tabela_newton_b = newton_metodo2(b, b_, fb, dfb, tolerancia, iteracao_max)
tabela_newton_c = newton_metodo2(c, c_, fc, dfc, tolerancia, iteracao_max)
tabela_newton_d = newton_metodo2(d, d_, fd, dfd, tolerancia, iteracao_max)

print("\nExercício 2) Use o método de bissecção para resolver os exemplos")
x_a,tabela_bisseccao_a = bisseccao_log(fa, 1, 4, tolerancia, iteracao_max)
x_b,tabela_bisseccao_b = bisseccao_log(fb, 0, PI/2, tolerancia, iteracao_max)
x_d,tabela_bisseccao_c = bisseccao_log(fc, -3, -2, tolerancia, iteracao_max)
x_d,tabela_bisseccao_d = bisseccao_log(fd, 0, PI/2, tolerancia, iteracao_max)

print("\nExercício 3) Compare o número de iterações utilizados entre os dois métodos, e apresente uma tabela com os resultados.")
print("Iterações estão contidas em vetores. [a iteração máxima escolhida foi 120]\n")
tabela_comp_iteracao = {
    'Comparativo':['Exercício a)','Exercício b)','Exercício C)','Exercício d)'],
    'Iterações feitas no Método bissecção':[tabela_bisseccao_a['Iteração'],tabela_bisseccao_b['Iteração'],tabela_bisseccao_c['Iteração'],tabela_bisseccao_d['Iteração']],
    'Iterações feitas no Método de Newton':[tabela_newton_a['Iteração'],tabela_newton_b['Iteração'],tabela_newton_c['Iteração'],tabela_newton_d['Iteração']],
}

print(tabulate(tabela_comp_iteracao, headers="keys", tablefmt='fancy_grid'))

```
